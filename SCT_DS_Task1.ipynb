```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'India_Age_Demographics.csv'  # Replace with the correct file path
data = pd.read_csv(file_path)
age_group_counts = data['Age group'].value_counts()



# Create a barplot
plt.figure(figsize=(10, 6))
plt.bar(age_group_counts.index, age_group_counts.values, color='blue', edgecolor='black')
plt.title('Age Distribution in India')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.80)
plt.xticks(rotation=45, horizontalalignment='right') # Rotate x-axis labels for better readability


# Show the plot
plt.show()
```
