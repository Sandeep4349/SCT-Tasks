```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Titanic.csv')
print("Data loaded successfully.")

    # --- Data Cleaning ---

# 1. Check for missing values
print("\nMissing Values:")
print(df.isna().sum())



    #2. Check for duplicates (and remove if needed)
print("\nDuplicate Rows:")
print(df.duplicated().sum())
    # df.drop_duplicates(inplace=True)  # Uncomment to remove duplicates

    # 3. Feature Engineering (creating new useful features)
    # Example: Create 'FamilySize'
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1  # +1 for the passenger themselves

    # Example: Create 'IsAlone'
df['IsAlone'] = 0
df.loc[df['FamilySize'] == 1, 'IsAlone'] = 1
print(df['IsAlone'])

    # 4. Convert categorical features to numerical (if needed for modeling later)
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
print(df['Sex'],df['Embarked'])

    # --- EDA (Exploratory Data Analysis) ---

    # 1. Descriptive statistics
print("\nDescriptive Statistics:")
print(df.describe())

    # 2. Univariate Analysis (distributions of individual variables)
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution')
plt.show()

print('----------')

sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()

print('----------')

    # 3. Bivariate Analysis (relationships between variables)
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Age', data=df)  # Age vs. Survival
plt.title('Age vs. Survival')
plt.show()

print("-----------")

plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='Sex', data=df)  # Survival by Gender
plt.title('Survival by Gender')
plt.show()

print('------------')

plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Survived', data=df)  # Survival by Pclass
plt.title('Survival by Pclass')
plt.show()

print('------------')
    # 4. Correlation Matrix

numerical_df = df.select_dtypes(include=['number'])
correlation_matrix = numerical_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

print('------------------')
print(df.columns)
print(df.loc[[1,4]])
print(df[1:4])
print(df.iloc[[1,2]])

```
